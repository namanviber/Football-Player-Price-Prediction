
Since there are so many categorical values, we need to use feature selection We need to perform label encoding before applying chi 2 analysis



```python
#importing label encoing module
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

#creating a new data frame from performing the chi2 analysis
chi2data=pd.DataFrame()

#adding all the categorical columns except the output to new data frame
for i in categorical:
    if i!= 'Accident_severity':
        chi2data[i]=le.fit_transform(trainset[i])
```

```python
chi2data.info()
```

```python
plt.figure(figsize=(22,17))
sns.set(font_scale=1)
sns.heatmap(chi2data.corr(), annot=True)
```

```python
#label encoded data set
chi2data.head(10)
```

```python
#for better understanding and ease of access adding them to a new dataframe
from sklearn.feature_selection import chi2

f_p_values=chi2(chi2data,trainset['Accident_severity'])
f_p_values1=pd.DataFrame({'features':chi2data.columns, 'Fscore': f_p_values[0], 'Pvalues':f_p_values[1]})
f_p_values1.sort_values(by='Pvalues',ascending=True)
```


We need higher Fscore and lower the Pvalues, so by evaluating, we can remove 'Pedestrian_movement',
        'Educational_level','Sex_of_driver'

```python
#after evaluating we are removing lesser important columns and storing to a new data frame
data=trainset.drop([  'Pedestrian_movement',
        'Educational_level','Sex_of_driver'
         ],axis=1)
data.head()
```

```python
print("Shape of Data: ", data.shape)
data.info
```

```python
#to check distinct values in each categorical columns we are storing them to a new variable
data_categorical=[i for i in data.columns if data[i].dtype=='O']
data_categorical
```

```python
for i in data_categorical:
    print(data[i].value_counts())
```

```python
# Get dummy for Encoding Categorical data
data_dummy=pd.get_dummies(data[data_categorical],drop_first=True)
data_dummy.head()
```

```python
#concatinate dummy and data frame
concatinatedData=pd.concat([data,data_dummy],axis=1)
concatinatedData.head()
```

```python
#dropping dummied columns to get label encoded dataframe
concatinatedData.drop(data_categorical,axis=1,inplace=True)
concatinatedData.head()
```

Seperating Input Columns and Output COlumns

```python
input_variables = concatinatedData.drop(['Accident_severity'],axis=1)
print("Dimensions of input variables Dataframe: ",input_variables.shape)
input_variables.head()
```

```python
output_variables = concatinatedData['Accident_severity']
print("Dimensions of output variables Dataframe: ",output_variables.shape)
output_variables.head()
```

```python
#checking the count of Severity of Accidents
output_variables.value_counts()
```

```python
#plotting count plot using seaborn
sns.countplot(x = output_variables, palette='muted')
```

Oversampling Data to improve robustness of model and removing noise

```python
#importing SMOTE 
from imblearn.over_sampling import SMOTE
oversample=SMOTE()
xo,yo=oversample.fit_resample(input_variables,output_variables)
yo.value_counts()
```

```python
#converting data to training data and testing data

from sklearn.model_selection import train_test_split

#splitting 75% of the data to training data and 25% of data to testing data

x_train,x_test,y_train,y_test=train_test_split(xo,yo,test_size=0.25,random_state=42)
```

```python
print("x_train: ",x_train.shape)
print("x_test: ",x_test.shape)
print("y_train: ",y_train.shape)
print("y_test: ",y_test.shape)
```

<font size="24">Building Models</font>

KNN Model Algorithm

```python
# import the necessary liabrary
from sklearn.neighbors import KNeighborsClassifier

# modelling using random forest baseline
model_KNN=KNeighborsClassifier(n_neighbors=5)
model_KNN.fit(x_train,y_train)

# predicting on test data
predict_knn = model_KNN.predict(x_test)

# train score 
print("KNN Model Accuracy Score: ",model_KNN.score(x_train, y_train))
```

```python
#Importing Libraries
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,ConfusionMatrixDisplay

matrix_KNN=confusion_matrix(y_test,predict_knn)
print(matrix_KNN,'\n')
print(ConfusionMatrixDisplay.from_predictions(y_test,predict_knn))
accuracy_KNN=accuracy_score(y_test,predict_knn)
print(accuracy_KNN,'\n')
report_KNN=classification_report(y_test,predict_knn)
print(report_KNN)
```
